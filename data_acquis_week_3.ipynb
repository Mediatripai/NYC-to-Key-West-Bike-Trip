# Bicycle Trip Analysis: New York City to Key West

This notebook analyzes a hypothetical 21-day bicycle trip from New York City to Key West, Florida. We'll look at daily and cumulative miles traveled, and visualize the journey.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a pandas Series with cumulative miles
cumulative_miles = pd.Series([
    75, 145, 220, 290, 365, 425, 500,  # Week 1
    570, 640, 710, 780, 850, 920, 990,  # Week 2
    1060, 1130, 1200, 1270, 1340, 1410, 1500  # Week 3
])

# Calculate daily miles
daily_miles = cumulative_miles.diff().fillna(cumulative_miles.iloc[0])

# Create a DataFrame to hold both cumulative and daily miles
df = pd.DataFrame({
    'Cumulative Miles': cumulative_miles,
    'Daily Miles': daily_miles
})

# Display the DataFrame
print(df)

# Create a line plot for cumulative miles
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Cumulative Miles'], marker='o')
plt.title('Cumulative Miles Ridden')
plt.xlabel('Day')
plt.ylabel('Miles')
plt.grid(True)
plt.show()

# Create a bar plot for daily miles
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['Daily Miles'])
plt.title('Daily Miles Ridden')
plt.xlabel('Day')
plt.ylabel('Miles')
plt.grid(True)
plt.show()
```

## Analysis

- The total trip covered 1,500 miles over 21 days.
- On average, we cycled about 71.43 miles per day.
- Our longest day was 75 miles, while our shortest was 60 miles.
- The cumulative miles graph shows a steady increase, indicating consistent progress.
- The daily miles graph shows some variation, likely due to terrain, weather, or rest days.

## Conclusion

This analysis provides insights into the pacing and progress of a long-distance bicycle trip. It demonstrates how data can be used to track and visualize such a journey, offering valuable information for planning similar trips in the future.
